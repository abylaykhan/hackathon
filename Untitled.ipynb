```python
import pandas as pd
import numpy as np
from datetime import timedelta

# Загрузим данные (предполагаем, что они в DataFrame df)
# df['user_creation_time'] = pd.to_datetime(df['user_creation_time'])

# Признак 1: Время с момента создания пользователя до события
# df['time_since_creation'] = (df['event_time'] - df['user_creation_time']).dt.total_seconds()

# Признак 2: Частота событий пользователя за последние 24 часа
df['event_time_24h'] = df.groupby('user_id')['event_time'].transform(
    lambda x: x.apply(lambda t: (df['event_time'] - t).dt.total_seconds() <= 24 * 60 * 60).sum()
)

# Признак 3: Частота событий по session_id
df['session_event_count'] = df.groupby('session_id')['event_id'].transform('count')

# Признак 4: Проверка аномальной активности по IP-адресу (если есть подозрительные значения)
df['ip_country'] = df['ip_address'].apply(lambda x: x.split('.')[0])  # Пример извлечения страны по IP

# Признак 5: Количество уникальных устройств пользователя
df['unique_devices'] = df.groupby('user_id')['device_id'].transform('nunique')

# Признак 6: Признаки для геолокации (например, разница в локации)
df['location_shift'] = df.groupby('user_id').apply(
    lambda group: np.abs(group['location_lat'].diff()) + np.abs(group['location_lng'].diff())
).fillna(0)

# Признак 7: Признаки на основе event_type (например, количество уникальных типов событий)
df['unique_event_types'] = df.groupby('user_id')['event_type'].transform('nunique')

# Признак 8: Количество событий по платформам и ОС
df['platform_event_count'] = df.groupby('platform')['event_id'].transform('count')
df['os_event_count'] = df.groupby('os_name')['event_id'].transform('count')

# Признак 9: Время события в день недели (выходные/рабочие дни)
df['event_day_of_week'] = df['event_time'].dt.dayofweek

# Признак 10: Разница во времени между двумя последовательными событиями
df['time_diff_prev_event'] = df.groupby('user_id')['event_time'].diff().dt.total_seconds().fillna(0)

# Признак 11: Повторяемость действий (сколько раз был тот же тип события подряд)
df['event_type_repeat_count'] = df.groupby('user_id')['event_type'].transform(lambda x: x.duplicated(keep=False).sum())

# Признак 12: Средняя длина сессии (в секундах)
df['session_duration'] = df.groupby('session_id')['event_time'].transform(lambda x: (x.max() - x.min()).total_seconds())

# Признак 13: Процент изменений устройства между событиями
df['device_change_pct'] = df.groupby('user_id')['device_id'].transform(lambda x: (x != x.shift()).mean())
```
